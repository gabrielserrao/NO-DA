```python

M = pd.read_pickle(f'/samoa/data/smrserraoseabr/NO-DA/historymatching/ESMDA/filterdata_REF1000_PARALLEL_simulations_Loc_None_PERM_Nit8_HF100_PX0/it3/MGrid_3.pkl')
M_Prior = pd.read_pickle('/samoa/data/smrserraoseabr/NO-DA/historymatching/ESMDA/filterdata_REF1000_PARALLEL_simulations_Loc_None_PERM_Nit8_HF100_PX0/it0/MGrid_0.pkl')
M= np.exp(np.array(M).reshape(32,32,100))
M_Prior= np.exp(np.array(M_Prior).reshape(32,32,100))
#no exp
#M= np.array(M).reshape(32,32,100)
#M_Prior= np.array(M_Prior).reshape(32,32,100)
```

```python
#plot M and M_Prior side by side
sample = 50
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.imshow(M[:,:,sample], cmap='jet')
plt.colorbar()
plt.title('M')
plt.subplot(1,2,2)
plt.imshow(M_Prior[:,:,sample], cmap='jet')
plt.colorbar()
plt.title('M_prior')
plt.show()


#plot the histogram bello each map
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(M[:,:,sample].flatten(), bins=50)
plt.title('M')
plt.subplot(1,2,2)
plt.hist(M_Prior[:,:,sample].flatten(), bins=50)
plt.title('M_prior')
plt.show()

#print the statistics of each map
print('M')
print('mean: ', np.mean(M[:,:,sample]))
print('std: ', np.std(M[:,:,sample]))
print('M_prior')
print('mean: ', np.mean(M_Prior[:,:,sample]))
print('std: ', np.std(M_Prior[:,:,sample]))

```

```python
from scipy.stats import boxcox

# Assuming your data is in a one-dimensional array or list named "data"
data_transformed, lambda_best_fit = boxcox(M.flatten())

# data_transformed now contains the transformed data
# lambda_best_fit is the best-fit value for lambda

#plot the histogram of the transformed data
plt.hist(data_transformed, bins=20)
plt.show()
```

```python
from scipy.special import inv_boxcox

# Using the transformed data and the lambda value found earlier
data_recovered = inv_boxcox(data_transformed, lambda_best_fit)
```

```python
#compare the M hist with the data_recovered hist

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(data_recovered,bins=50,label='data_recovered')
plt.legend()
plt.subplot(1,2,2)
plt.hist(M,bins=50,label='M')
plt.legend()




```
